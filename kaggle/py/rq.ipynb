```python
import pandas as pd

class LanguageData:
    def __init__(self, data_source):
        if isinstance(data_source, str):
            self.data = pd.read_csv(data_source)
        else:
            self.data = data_source

        self.number_of_years = self.data['year'].nunique()
        self.number_of_countries = self.data['iso2_code'].nunique()
        self.number_of_unique_quarters = self.data['quarter'].nunique()

    def get_programmers(self, country_code, language, year, quarter):
        """Get number of programmers for specific country, language, year, quarter"""
        result = self.data[
            (self.data['iso2_code'] == country_code) &
            (self.data['language'] == language) &
            (self.data['year'] == year) &
            (self.data['quarter'] == quarter)
        ]
        return result['num_pushers'].iloc[0] if not result.empty else None


# Uso correcto
url = "https://raw.githubusercontent.com/github/innovationgraph/main/data/languages.csv"
analyzer = LanguageData(url)

print(f"Años: {analyzer.number_of_years}")
print(f"Países: {analyzer.number_of_countries}")
print(f"Trimestres: {analyzer.number_of_unique_quarters}")

# Ejemplo: HTML en Perú 2022 Q2
result = analyzer.get_programmers("PE", "HTML", 2022, 2)
print(f"Programadores HTML en Perú 2022 Q2: {result}")
```
