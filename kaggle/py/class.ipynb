```python
class Coche:
    def __init__(self, marca, modelo):
        self.marca = marca
        self.modelo = modelo
        self.velocidad = 0  # velocidad inicial
    
    def acelerar(self):
        self.velocidad += 10
        print(f"El coche {self.marca} {self.modelo} acelera a {self.velocidad} km/h")

    def frenar(self):
        if self.velocidad > 0:
            self.velocidad -= 5
        print(f"El coche {self.marca} {self.modelo} frena a {self.velocidad} km/h")

# Probar
mi_coche = Coche("Honda", "Civic")
mi_coche.acelerar()
mi_coche.acelerar()
mi_coche.frenar()
mi_coche.acelerar()
```
